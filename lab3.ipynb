One way ANOVA
Since we only have one dataset, we can use it to create multiple samples to perform both methods of ANOVA
We will create 3 datasets of 6 rows each and then perform our tests
For One Way ANOVA we will use qsec column of the dataframe, assuming α=0.05
NULL HYPOTHESIS=> All 3 samples belong to the same population, ie, mu1=mu2=mu3
ALTERANTE HYPOTHESIS=> Not all samples belong to the same population, or, at least 2 of the samples have diff means

```r
set.seed(5)
sample1 = data[sample(nrow(data),6),]
sample1
```

```r
set.seed(8)
sample2 = data[sample(nrow(data),6),]
sample2
```

```r
set.seed(12)
sample3 = data[sample(nrow(data),6),]
sample3
```

```r
sample1$group<-"sample1"
sample1
sample2$group<-"sample2"
sample2
sample3$group<-"sample3"
sample3
```

```r
sample<-rbind(sample1,sample2,sample3)
sample
```

```r
res.aov<-aov(qsec~group,data=sample)
summary(res.aov)
```

Since P value(0.245)>alpha(0.05)
Therfore,We accept the null hypothesis that all three samples belong to same polulation

```r
anova<-function(data,col,group){
 stats<-aggregate(col, list(group), FUN=mean)
 means<-stats[,2]
 cnt<-count(data,group)[,2]
 k<-nrow(stats)
 mean_total=sum(col)/nrow(data)
 dof1=k-1
 dof2=nrow(data)-k
 bss<-0
 for(i in 1:k){
 bss=bss+cnt[i]*(means[i]-mean_total)**2
 }
 msb=bss/dof1
 wss<-0
 for(i in 1:nrow(data)){
 curr_mean=means[stats[1]==group[i]]
 wss=wss+(col[i]-curr_mean)**2
 }
 msw=wss/dof2
 fcal=msb/msw
 return(fcal)
}
```

```r
anova(sample,sample$qsec,sample$group)
```

Two way ANOVA
We will use the same sample dataframe that we built earlier using random values
This time we will be using column-disp against am and group columns
NULL HYPOTHESIS A=>Mean of disp is same for all groups
ALTERANTE HYPOTHESIS A=>At least two of the groups have different means
NULL HYPOTHESIS B=>Mean of disp is same for both values of am
ALTERANTE HYPOTHESIS=>Both values of am have different means

```r
aov2<-aov(disp~group+am,data=sample)
```

```r
summary(aov2)
```

From first p value(0.69>0.05) We can conclude that the mean for all three groups is same
From second p value(0.0004<0.05) We can reject null hypothesis B, so both values of am have different means

```r
two_way_anova<-function(data,col,group1,group2){
 statsy<-aggregate(col,list(group1),FUN=mean)
 statsx<-aggregate(col,list(group2),FUN=mean)
 meansy<-statsy[,2]
 meansx<-statsx[,2]
 cnty<-aggregate(col,list(group1),FUN=length)[,2]
 cntx<-aggregate(col,list(group2),FUN=length)[,2]
 grand_mean=sum(col)/nrow(data)
 p=nrow(statsy)
 q=nrow(statsx)
 ssa=q*(sum((meansy-grand_mean)**2))*3
 ssb=p*(sum((meansx-grand_mean)**2))*3
 sse<-0
 for(i in 1:nrow(data)){
 curr_meany=meansy[statsy[1]==group1[i]]
 curr_meanx=meansx[statsx[1]==group2[i]]
 sse=sse+((col[i]-curr_meany-curr_meanx+grand_mean)**2)
 }
 msa=ssa/(p-1)
 msb=ssb/(q-1)
 mse=sse/(cnty[1]+cntx[1]-1)
 f1=msa/mse
 f2=msb/mse
 return(list(f1,f2))
}
```

```r
two_way_anova(sample,sample$disp,sample$group,sample$am)
```

We get the same f values as we saw earlier,Hence the same conclusion
